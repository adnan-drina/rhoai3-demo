# Detect Faces and Draw Bounding Boxes

In this notebook we learn to **draw bounding boxes** around faces detected by
YOLO11 and inspect the raw detection results: confidence scores, pixel
coordinates, and class IDs.

```python
!pip install --no-cache-dir -q -r requirements.txt

from ultralytics import YOLO
from huggingface_hub import hf_hub_download
from PIL import Image
```

```python
model_path = hf_hub_download(repo_id="AdamCodd/YOLOv11n-face-detection", filename="model.pt")
model = YOLO(model_path)
```

```python
results = model.predict("images/test_face_01.jpg")
result = results[0]
```

```python
len(result.boxes)
```

```python
box = result.boxes[0]
print("Object type:", box.cls)
print("Coordinates:", box.xyxy)
print("Probability:", box.conf)
```

```python
cords = box.xyxy[0].tolist()
class_id = box.cls[0].item()
conf = box.conf[0].item()
print("Object type:", class_id)
print("Coordinates:", cords)
print("Probability:", conf)
```

```python
print(result.names)
```

## Understanding the class map

The pre-trained model has a **single class** — `face`.  Every detection is
labelled with class ID `0` → `"face"`.

After retraining in notebook 03, the model will have **two classes** — for
example `"adnan"` and `"unknown_face"` — allowing it to identify specific
individuals.

```python
Image.fromarray(result.plot()[:, :, ::-1])
```

```python
import glob

group_images = sorted(glob.glob("images/test_group_*.jpg"))
for img_path in group_images:
    print(f"\n=== {img_path} ===")
    results = model.predict(img_path)
    result = results[0]

    for box in result.boxes:
        class_id = result.names[box.cls[0].item()]
        cords = [round(x) for x in box.xyxy[0].tolist()]
        conf = round(box.conf[0].item(), 2)
        print(f"  {class_id}: {cords}  conf={conf}")

    display(Image.fromarray(result.plot()[:, :, ::-1]))
```
